```python
import pandas as pd

Apple_dataframe = pd.read_csv('AMZN_data.csv')
Apple_dataframe.head()
```

```python
open = []
high = []
low = []
close = []
dates = []
for i in range(len(Apple_dataframe)):
    date = Apple_dataframe['Date'][i].split(' ')[0]
    year = date[2:4]
    if year == '24':
        open.append(Apple_dataframe['Open'][i])
        high.append(Apple_dataframe['High'][i])
        low.append(Apple_dataframe['Low'][i])
        close.append(Apple_dataframe['Close'][i])
        dates.append(Apple_dataframe['Date'][i])
        

        
```

```python
apple_dict =  {"Date" : dates , "Open" : open , "High" : high , "Low" : low , "Close" : close}
apple_test = pd.DataFrame(apple_dict)
apple_test.to_csv('Amazon_test.csv')
```

```python
length = 40
test = pd.read_csv('Amazon_test.csv')
i  = 0
k = 0
while i < len(test):
    open_test = test['Open'][i:i+length]
    high_test = test['High'][i:i+length]
    low_test = test['Low'][i:i+length]
    close_test = test['Close'][i:i+length]
    test_df = pd.DataFrame({"Open" : open_test , "High" : high_test , "Low" : low_test , "Close" : close_test})
    test_df.to_csv(f"Amazon_test_set_{k}.csv")
    k += 1
    i += length

```
